```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for precision, recall, and F1 score for different batch sizes
batch_sizes = [16, 32, 64]
#50 epochs
precision_scores = [0.92509, 0.91875, 0.92242]  # Replace with your actual precision scores
recall_scores = [0.93730, 0.93594, 0.93541]     # Replace with your actual recall scores
f1_scores = [0.93110, 0.92716, 0.92880]         # Replace with your actual F1 scores
#20 epochs
# precision_scores = [0.91806, 0.92047, 0.92115]  # Replace with your actual precision scores
# recall_scores = [0.93142, 0.93399, 0.93156]     # Replace with your actual recall scores
# f1_scores = [0.92463, 0.92710, 0.92628]         # Replace with your actual F1 scores

# Create subplots for precision, recall, and F1 score
plt.figure(figsize=(10, 6))

plt.plot(batch_sizes, precision_scores, marker='o', label='Precision', linestyle='-', color='blue')
plt.plot(batch_sizes, recall_scores, marker='s', label='Recall', linestyle='--', color='green')
plt.plot(batch_sizes, f1_scores, marker='^', label='F1 Score', linestyle='-.', color='red')

plt.xlabel('Batch Size')
plt.ylabel('Score')
plt.title('Precision, Recall, and F1 Score vs. Batch Size (epoch=50;lr=1e-5)')
plt.legend(loc='best')
plt.grid(True)

plt.xticks(batch_sizes)
plt.tight_layout()

# Save the plot or display it
plt.show()
```
